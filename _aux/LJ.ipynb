```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(6,4), dpi=400)
plt.rcParams.update({'font.size': 14})
plt.rcParams.update({'pdf.fonttype': 42})

r = np.linspace(1, 3, 1000)
plt.plot(r, 1/r**12 - 1/r**6, label=r'$\dfrac{1}{r^{12}} - \dfrac{1}{r^6}$')
plt.legend()
plt.xlabel(r'$r$')

# plt.savefig('LJ_potential.pdf', bbox_inches='tight')
plt.savefig('LJ_potential.png', bbox_inches='tight')
plt.show()
```
